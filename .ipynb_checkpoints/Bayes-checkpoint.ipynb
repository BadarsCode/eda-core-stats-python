```python
# Bayes example from above
p_disease = 0.01
sensitivity = 0.99     # P(test+ | disease)
specificity = 0.95     # P(test- | no disease)

num = sensitivity * p_disease
den = num + (1 - specificity) * (1 - p_disease)
p_disease_given_positive = num / den
print(p_disease_given_positive)  # ~0.1666667
```

```python
import numpy as np
import matplotlib.pyplot as plt

# population: exponential (skewed)
pop = np.random.exponential(scale=1.0, size=1000000)

n = 30   # sample size per draw
trials = 2000
sample_means = [np.mean(np.random.choice(pop, n)) for _ in range(trials)]

plt.hist(sample_means, bins=30, edgecolor='k')
plt.title('Sampling distribution of the mean (n=30)')
plt.show()
```

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, norm, expon
```

```python
n, p = 10, 0.3
k_vals = np.arange(0, n+1)
binom_pmf = binom.pmf(k_vals, n, p)
```

```python
prob_binom_le2 = binom.cdf(2, n, p)
binom_mean, binom_var = binom.mean(n, p), binom.var(n, p)
```

```python
plt.figure(figsize=(8,4))
plt.bar(k_vals, binom_pmf, color="skyblue", edgecolor="black")
plt.title("Binomial PMF (n=10, p=0.3)")
plt.xlabel("k (successes)")
plt.ylabel("P(X=k)")
plt.grid(axis="y", alpha=0.3)
plt.show()
```

```python
# Poisson Distribution (Discrete)
lam = 3
k_vals_pois = np.arange(0, 15+1)
pois_pmf = poisson.pmf(k_vals_pois, lam)
```

```python
prob_pois_ge5 = 1 - poisson.cdf(4, lam)
pois_mean, pois_var = poisson.mean(lam), poisson.var(lam)
```

```python
plt.figure(figsize=(8,4))
plt.bar(k_vals_pois, pois_pmf, color="lightgreen", edgecolor="black")
plt.title("Poisson PMF (lambda=3)")
plt.xlabel("k (count)")
plt.ylabel("P(X=k)")
plt.grid(axis="y", alpha=0.3)
plt.show()
```

```python
# Normal Distribution (Continuous)
mu, sigma = 0, 1
x = np.linspace(-4, 4, 400)
normal_pdf = norm.pdf(x, mu, sigma)
```

```python
prob_norm_between = norm.cdf(1, mu, sigma) - norm.cdf(-1, mu, sigma)
norm_mean, norm_var = norm.mean(mu, sigma), norm.var(mu, sigma)
```

```python
plt.figure(figsize=(8,4))
plt.plot(x, normal_pdf, color="blue")
plt.title("Normal PDF (mu=0, sigma=1)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(alpha=0.3)
plt.show()
```

```python
# Exponential Distribution (Continuous)
rate = 1.0  # lambda
x_exp = np.linspace(0, 8, 400)
exp_pdf = expon.pdf(x_exp, scale=1/rate)
```

```python
prob_exp_gt2 = 1 - expon.cdf(2, scale=1/rate)
exp_mean, exp_var = expon.mean(scale=1/rate), expon.var(scale=1/rate)
```

```python
plt.figure(figsize=(8,4))
plt.plot(x_exp, exp_pdf, color="red")
plt.title("Exponential PDF (lambda=1)")
plt.xlabel("x (time)")
plt.ylabel("f(x)")
plt.grid(alpha=0.3)
plt.show()
```
