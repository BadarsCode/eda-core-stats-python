```python
# Bayes example from above
p_disease = 0.01
sensitivity = 0.99     # P(test+ | disease)
specificity = 0.95     # P(test- | no disease)

num = sensitivity * p_disease
den = num + (1 - specificity) * (1 - p_disease)
p_disease_given_positive = num / den
print(p_disease_given_positive)  # ~0.1666667
```

```python
import numpy as np
import matplotlib.pyplot as plt

# population: exponential (skewed)
pop = np.random.exponential(scale=1.0, size=1000000)

n = 30   # sample size per draw
trials = 2000
sample_means = [np.mean(np.random.choice(pop, n)) for _ in range(trials)]

plt.hist(sample_means, bins=30, edgecolor='k')
plt.title('Sampling distribution of the mean (n=30)')
plt.show()
```
